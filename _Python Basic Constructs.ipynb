Q19.Write a function which finds all pythagorean triplets of triangles whose sides are no greater than a natural number N.

```python
def find_pythagorean_triplets(N):
    triplets = []
    for a in range(1, N+1):
        for b in range(a, N+1):  # Start b from a to avoid duplicates
            c = (a**2 + b**2)**0.5
            if c.is_integer() and c <= N:  # Check if c is a perfect square and within the limit
                triplets.append((a, b, int(c)))
    return triplets

# Example usage
N = 10
triplets = find_pythagorean_triplets(N)
print("Pythagorean triplets with sides no greater than", N, "are:", triplets)
```
